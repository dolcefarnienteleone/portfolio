## Import necessary packages

```python
%matplotlib inline

# Some usual imports here
import csv as csv 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Import data for making "People" pillar

```python
census = pd.read_csv('2011_london_census.csv')
gcse = pd.ExcelFile('gcse-results-old.xls')
gcse = pd.read_excel(gcse,'2011-12', index_col=None,header=2)
lifeexpect = pd.ExcelFile('healthy-life-expectancy.xls')
malelife = pd.read_excel(lifeexpect,'Males',index_col=None)
femalelife = pd.read_excel(lifeexpect,'Females',index_col=None)
```

Checking data structure

```python
census.head()
#check the number of features
list(census.columns)
```

People pillar ---\
Group census data by Local Authority and pick the columns we need: a. age b. ethnic c. married or not
1. (a) age

```python
agecol = ['Persons aged 0 to 4','Persons aged 5 to 14','Persons aged 15 to 24','Persons aged 25 to 44','Persons aged 45 to 64','Persons aged 65 to 89','Persons aged 90 and over']
agecensus = census.groupby('Local Authority')[agecol].sum()
agecensus
```

People pillar --- \
1. (b) Ethnic

```python
ethniccol = ['Persons who are white',
 'Persons who have mixed ethnicity or are from multiple ethnic groups',
 'Persons who are Asian/Asian British: Indian',
 'Persons who are Asian/Asian British: Pakistani',
 'Persons who are Asian/Asian British: Bangladeshi',
 'Persons who are Asian/Asian British: Chinese and Other',
 'Persons who are Black/African/Caribbean/Black British',
 'Persons who are Arab or from other ethnic groups']
ethniccensus = census.groupby('Local Authority')[ethniccol].sum()
ethniccensus
```

People pillar ---\
1. (c) Engineer new features for Married or not: ('Persons aged over 16 who are married or in a registered same-sex civil partnership' divided by 'Total Population aged 16 and over')

```python
colneeds = ['Persons aged over 16 who are married or in a registered same-sex civil partnership','Total Population aged 16 and over']
marriedcensus = census.groupby('Local Authority')[colneeds].sum()
#adding new feature to marriedcensus by calculate married percentage
marriedcensus['marriedpercentage']=marriedcensus['Persons aged over 16 who are married or in a registered same-sex civil partnership']/marriedcensus['Total Population aged 16 and over']
marriedcensus
```

1. (d) Grade: there is missing data for City of London, need to impute the NaN

```python
#select the only column in need
gcse = gcse.iloc[1:34,0:10]
#use #9 Average GCSE and equivalents point score per pupil as the gcse indicator
gcse = gcse.iloc[:,[0,1,9]]
#since there is no score data for the first row (City of London), using mean to represent the score
m = gcse.iloc[2:34,2].mean(axis=0)
m
a = gcse.iloc[0,2]
gcse = gcse.replace(a,m)
#rename columns for next use to combine into other people indicators
gcse.columns = ['Code','Local Authority','Grade']
gcse
```

1. (e) Higher Edu indicator by engineering new feature: higher education people divided by total population withing the borough

```python
highedu = census[['Local Authority','Total Population aged 16 and over','Persons aged over 16 whose highest level of qualification is Level 4 qualifications and above']]
highedu = highedu.groupby('Local Authority').sum()
#calculate overall percentage of higher education
highedu['high edu percentage'] = highedu['Persons aged over 16 whose highest level of qualification is Level 4 qualifications and above']/highedu['Total Population aged 16 and over']
highedu
```

1. (f) Life Expectancy Average: the dataset originally separates female and male, therefore, need to create a new feature with the average of both sexual life expectancy

```python
#life expectancy at birth indicator (by male and female will need transform to avg per person)
malelife = malelife.iloc[2:34,[0,1,3]]
malelife.columns = ['Code','Local Authority','malelife']
femalelife = femalelife.iloc[2:34,[0,1,3]]
femalelife.columns = ['Code','Local Authority','femalelife']
#combine male & female life expectancy to get the avg for whole
lifelong = pd.merge(malelife,femalelife,on='Local Authority',how='outer')
#eliminate duplicate LOA code
lifelong = lifelong.iloc[:,[0,1,2,4]]
#adding mean of all
lifelong['life expectancy avg'] = (lifelong['malelife']+lifelong['femalelife'])/2
lifelong
```

### Combine all the indicators for 1st pillar - People

```python
#age & ethinic group indicator combine
people = pd.merge(agecensus, ethniccensus, on='Local Authority',how='outer')
#adding married indicator
people = pd.merge(people, marriedcensus, on='Local Authority',how='outer')
#adding life expectancy indicator
people = pd.merge(lifelong,people,on='Local Authority',how='outer')
#adding highedu to people pillar
people = pd.merge(people,highedu,on='Local Authority')
#adding gcse score to people pillar
people = pd.merge(people,gcse, on='Local Authority',how='outer')
people
```

Final cleaning for the people pillar indicators

```python
list(people.columns)
peopledropcols = ['malelife','femalelife',
                  'Persons aged over 16 who are married or in a registered same-sex civil partnership',
                  'Total Population aged 16 and over_x','Total Population aged 16 and over_y',
                  'Persons aged over 16 whose highest level of qualification is Level 4 qualifications and above',
                  'Code_x']
people = people.drop(peopledropcols,axis=1)
#change non numeric data to numeric data (object to float64)
people['life expectancy avg'] = people['life expectancy avg'].apply(pd.to_numeric,errors='coerce')
people.dtypes
#filling NaN with mean since all the NaN is from City of London (the original data collection is no data)
people = people.fillna(people.mean())
people
```

## Making 2nd pillar - Planet
### 2. Planet - a) Green+Blue Cover, b) Access to Nature, c) Air Pollution(NO2), d) GHG Emission

```python
#import data
gbcover = pd.ExcelFile('green_cover_borough_summary_0.05.xlsx')
gbcover = pd.read_excel(gbcover,'borough_green_cover', index_col=None)
accessN = pd.read_csv('public-open-space-nature-ward_access-to-nature.csv', index_col=None, header=0)
NO2 = pd.ExcelFile('ListofSchools_LAEI2013.xlsx')
NO2 = pd.read_excel(NO2,'List of schools', index_col=None,header = 3)
GHG = pd.read_csv('co2-emissions-borough-leggi.csv')
```

2. (a) GB cover indicator

```python
gbcover = gbcover.iloc[:,[0,1,8]]
gbcover.columns = ['Local Authority','Code','GB cover percent']
gbcover
```

2. (b) Access to nature indicator - group by Borough and calculate the average

```python
accessN = accessN.groupby('Borough Code')['% homes with good access to nature'].mean()
accessN
```

2. (c) NO2 (Air Pollution) indicator - grouping and use average

```python
NO2 = NO2.groupby('Local Authority name')['NO2ug/m3 mean 2013'].mean()
NO2
```

2. (d) GHG indicator - locate year 2011 and use the total emission

```python
GHG2011 = GHG.loc[(GHG['LEGGI_Year'] == 2011) & (GHG['Sector'] == 'Total')]
GHG2011 = GHG2011[['Borough','KtCO2e']]
GHG2011.columns = ['Local Authority','KtCO2e']
#reset index and drop the old index
GHG2011 = GHG2011.reset_index(drop=True)
#delete the last two rows since they are not borough data
GHG2011 = GHG2011[:-2]
GHG2011
```

#### Combine all the indicators for 2nd pillar - Planet

```python
#combine gbcover and accessN
planet = pd.merge(gbcover,accessN,left_on='Code', right_index=True, how='inner')
#combine planet with GHG2011
planet = pd.merge(planet,GHG2011, on='Local Authority', how='outer')
#since there are some different description on the same Locan Authority, e.g., Kingston upon Thames (Kinston), need to fill in the NAN value into correct cells
planet.iloc[20,4] = planet.iloc[33,4]
planet.iloc[26,4] = planet.iloc[34,4]
planet.iloc[32,4] = planet.iloc[35,4]
#delete the duplicate rows
planet = planet.iloc[0:-3]
#change the 'City of Westminster' to 'Westminster' to avoid future combination issue
planet.iloc[32,0] = 'Westminster'
#combine planet with NO2
planet = pd.merge(planet,NO2,left_on='Local Authority',right_index=True, how='outer')
#change non numeric data in to numeric form (object to float)
planet['KtCO2e'] = planet['KtCO2e'].apply(pd.to_numeric,errors='coerce')
planet.dtypes
planet
```

## Making 3rd pillar - Profit
### 3. Profit - a) Household Income, b) Housing Affordability, c) Unemployment rate, d) Tourism Spend Estimate, e) Tourist Ratio

```python
#import data
income = pd.read_csv('modelled-household-income-estimates-borough.csv')
afford = pd.ExcelFile('ratio-house-price-earnings-residence-based.xls')
afford = pd.read_excel(afford,'Median Earnings to Prices ratio', index_col=None)
unem = pd.ExcelFile('unemployment_in_london_2012.xls')
unem = pd.read_excel(unem,'Table 6',index_col=None,header=2)
trspend = pd.ExcelFile('tourism-spend-borough.xls')
trspend = pd.read_excel(trspend, '2009',index_col=None) #since there is no 2011 data on datastore, use 2009 as the indicator
tourratio = pd.ExcelFile('daytime-population-borough.xls')
tourratio = pd.read_excel(tourratio,'2014',index_col=None, header=1) #need to calculate the percentage
```

3. (a) Income indicator - locate 2011(2010/11) & use median data

```python
income = income.loc[(income['Measure'] == 'Median') & (income['Year'] == '2010/11')]
#reset index and drop the data that's not by borough
income = income.reset_index(drop=True)
income = income[:-15]
income.columns = ['Code','Local Authority','Measure','Year','Income']
income
```

3. (b) Affordability indicator - locate 2011

```python
afford = afford[['New Code','Area',2011]]
#drop the NAN data in the first row
afford = afford.drop([0])
afford = afford.reset_index(drop=True)
afford = afford.iloc[0:33,:]
afford.columns = ['Code','Local Authority','2011 Affordability']
afford
```

3. (c) unemployment rate indicator - locate 2011 (2010-2012) data

```python
unem.columns = ['Local Authority','Unemployment Level','CI (+/-)','Unemployment Rate','CI (+/-).1']
unem = unem[['Local Authority','Unemployment Rate']]
unem = unem.iloc[0:-4]
unem
```

3. (d) Tourism Spend Estimate indicator - sum up for overall expense

```python
trspend['All visitors (£ millions)'] = trspend['Overseas visitors (£ millions)']+trspend['Domestic visitors (£ millions)']+trspend['Day visitors (£ millions)']
trspend = trspend.drop([0]).reset_index()
trspend = trspend.iloc[0:-2]
trspend = trspend[['Area','All visitors (£ millions)']]
trspend.columns = ['Local Authority','All visitors (£ millions)']
trspend
```

3. (e) Tourist Ratio - generate new features by using toursits divided by workday population(excludes tourists))

```python
tourratio = tourratio.drop([0]).reset_index(drop=True)
list(tourratio.columns)
tourcols = ['Code','Boroughs','Workday Population (excludes tourists)',
            'Overseas Staying visitors','Domestic Staying Visitors','Day Trip Visitors ']
tourratio = tourratio[(tourcols)]
#the original dataframe dtype is object, transform into numbers
tourcolnum = ['Workday Population (excludes tourists)',
            'Overseas Staying visitors','Domestic Staying Visitors','Day Trip Visitors ']
tourratio[tourcolnum] = tourratio[tourcolnum].apply(pd.to_numeric, errors='coerce', axis=1)
#Adding new feature of Tourists Ratio by dividing the total tourists by workday population (excludes tourists)
tourratio['Tourists Ratio'] = (tourratio['Overseas Staying visitors']+tourratio['Domestic Staying Visitors']+tourratio['Day Trip Visitors '])/tourratio['Workday Population (excludes tourists)']
tourratio = tourratio.iloc[0:33]
tourratio = tourratio[['Code','Boroughs','Tourists Ratio']]
tourratio.columns = ['Code','Local Authority','Tourists Ratio']
tourratio
```

Combine all the indicators for 3rd pillar - Profit

```python
#combine income and afford
profit = pd.merge(income,afford,on='Code',how='outer')
profit = profit[['Code','Local Authority_x','Income','2011 Affordability']]
profit.columns = ['Code','Local Authority','Income','2011 Affordability']
#combine profit with unem
profit = pd.merge(profit,unem,on='Local Authority',how='outer')
#combine profit with trspend
profit = pd.merge(profit,trspend,on='Local Authority',how='outer')
#since there are some different description on the same Locan Authority, e.g., Kingston upon Thames(Kingston-upon-Thames), need to fill in NaN with correct data 
profit.iloc[20,5] = profit.iloc[33,5]
profit.iloc[26,5] = profit.iloc[34,5]
profit.iloc[32,5] = profit.iloc[35,5]
#delete the duplicate rows
profit = profit.iloc[0:-3]
#combine profit with tourratio
profit = pd.merge(profit,tourratio,on='Code',how='outer')
#remove duplicate Local Authority columns and rename
profit = profit[['Code',
 'Local Authority_x',
 'Income',
 '2011 Affordability',
 'Unemployment Rate',
 'All visitors (£ millions)',
 'Tourists Ratio']]
profit.columns = ['Code',
 'Local Authority',
 'Income',
 '2011 Affordability',
 'Unemployment Rate',
 'All visitors (£ millions)',
 'Tourists Ratio']
#filling NaN with mean since all the NaN is from City of London (the original data collection is no data)
profit = profit.fillna(profit.mean())
profit
```

```python
# Make a copy
corr1=corrsmallall.copy()

#In the correlation matrix, each *row* has the variable name
#as the index, but we need this to be a new column
corr1.index.name = 'var1'
corr1.reset_index(inplace=True)

#"melt" the table, preserving the column we've just made.
corr_pair=corr1.melt(id_vars=['var1'])
#replace the some of the column names with more sensible ones
corr_pair=corr_pair.rename(columns={'variable': 'var2','value': 'corr'})
#remove rows where var1 and var2 are the same (these are always 1)
corr_pair = corr_pair.drop(corr_pair[corr_pair['var1']==corr_pair['var2']].index)
```

```python
#make a histogram
plt.hist(corr_pair['corr'],20)
plt.title("Distributions of correlations between all pairs of numerical variables")

#sort the rows on the correlation
corr_pair=corr_pair.sort_values(by ='corr',ascending=False)

print()
print("Top ten most positively correlated pairs of variables")
print(corr_pair.head(10))

print()
print("Top ten most negatively correlated pairs of variables (in reverse order)")
print(corr_pair.tail(10))

#Add a column with the absolute correlation, so we can sort on
#amount of correlation, regardless of the sign if we want
corr_pair['abs_corr']=abs(corr_pair['corr'])
#Sort on this
corr_pair=corr_pair.sort_values(by ='abs_corr',ascending=False)
```

```python
print()
print("Top ten most correlated pairs of variables with '2011 Avg Happiness'")
print(corr_pair[corr_pair['var1']=='2011 Avg Happiness'].head(10))
```
